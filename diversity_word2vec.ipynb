```python
import MeCab
import unicodedata
import word2vec
from collections import Counter
```

```python
mt = MeCab.Tagger("-Owakati -d /usr/local/lib/mecab/dic/mecab-ipadic-neologd")
```

```python
#テキストデータ読みこみ
f = open('h27practice.txt')
input_txt = f.read()
f.close()
```

```python
#MACでPDFからのテキストデータを抽出した場合、正規化形式がNFDになってるケースがある
#そのため、分かち書きの際に濁点が正しく扱われないので、NFCに変換
input_txt = unicodedata.normalize("NFC", input_txt)

#Stop the wordsを削除
input_txt = input_txt.replace("○","")
input_txt = input_txt.replace("★","")
input_txt = input_txt.replace("・","")
input_txt = input_txt.replace("」","")
input_txt = input_txt.replace("「","")
input_txt = input_txt.replace("。","")
input_txt = input_txt.replace("、","")
input_txt = input_txt.replace(".","")
input_txt = input_txt.replace(" ","")
input_txt = input_txt.replace(",","")
input_txt = input_txt.replace("(","")
input_txt = input_txt.replace(")","")
input_txt = input_txt.replace("【", "")
input_txt = input_txt.replace("】","")
input_txt = input_txt.replace("▲","")
input_txt = input_txt.replace("/","")
```

```python
#分かち書き
wakati_txt = mt.parse(input_txt)
print(wakati_txt)
```

```python
#分かち書き結果の書き出し
f = open("wakati_txt.txt","w")
f.write(wakati_txt)
f.close()
```

```python
#word2vecの学習
word2vec.word2vec('wakati_txt.txt', 'wakati_txt.bin', size=300, verbose=True)
model = word2vec.load('wakati_txt.bin')
```

```python
indexes, metrics = model.cosine('ダイバーシティ')
model.generate_response(indexes, metrics).tolist()
```
